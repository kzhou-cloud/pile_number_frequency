```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open("int_counts.json") as f:
    int_counts = json.load(f)
with open("float_counts.json") as f:
    float_counts = json.load(f)

print(f"Unique integers: {len(int_counts):,}")
print(f"Unique floats:   {len(float_counts):,}")
print(f"Total int occurrences:   {sum(c for _, c in int_counts):,}")
print(f"Total float occurrences: {sum(c for _, c in float_counts):,}")
```

```python
# Top 30 integers
labels, counts = zip(*int_counts[:30])
fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(range(len(labels)), counts)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_ylabel("Count")
ax.set_title("Top 30 Integers in The Pile")
plt.tight_layout()
plt.show()
```

```python
# Top 30 floats
labels, counts = zip(*float_counts[:30])
fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(range(len(labels)), counts)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_ylabel("Count")
ax.set_title("Top 30 Floats in The Pile")
plt.tight_layout()
plt.show()
```
